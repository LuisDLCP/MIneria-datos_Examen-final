## 2) Identify outliers
Replace outliers by the mean value

```python
import numpy as np
```

```python
def replace(group):
    mean, std = group.mean(), group.std()
    outliers = (group - mean).abs() > 3*std
    group[outliers] = mean        # or "group[~outliers].mean()"
    return group

df_new = df.groupby('clase').transform(replace)
df_id = df["id"]
df_clase = df["clase"]
df2 = pd.concat([df_id, df_new, df_clase], axis=1, sort=False)
df2.head()
```

```python
df2.info()
```

## 3) Oversampling

```python
max_size = df2['clase'].value_counts().max()
max_size
```

```python
lst = [df2]
for class_index, group in df2.groupby('clase'):
    lst.append(group.sample(max_size-len(group), replace=True))
df3 = pd.concat(lst)
df3.sort_index(inplace=True)
df3.head()
```
